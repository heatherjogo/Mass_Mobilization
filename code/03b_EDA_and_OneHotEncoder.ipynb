### OneHotEncode country, region, stateresponses & protesterdemands columns

```python
# df with categorical variables that we want to encode
df = mass[['country', 'region','protesterdemand1', 'protesterdemand2', 'protesterdemand3',
       'protesterdemand4', 'stateresponse1', 'stateresponse2',
       'stateresponse3', 'stateresponse4', 'stateresponse5', 'stateresponse6',
       'stateresponse7', 'protest_size_category', 'protester_id_type']]
df.head()
```

```python
df.isnull().sum()
```

```python
# fill null values with 'None'
df.fillna('None', inplace=True)
df.isnull().sum()
```

```python
# one hot encode stateresponses
onehot = OneHotEncoder(dtype=np.int, sparse=True)

ohe_df = pd.DataFrame(onehot.fit_transform(df[['country', 'region','stateresponse1', 'stateresponse2', 'stateresponse3', 'stateresponse4',
       'stateresponse5', 'stateresponse6', 'stateresponse7', 'protesterdemand1',
       'protesterdemand2', 'protesterdemand3', 'protesterdemand4', 'protest_size_category', 'protester_id_type']]).toarray(),
                         columns=onehot.get_feature_names(['country', 'region','stateresponse1', 'stateresponse2', 'stateresponse3', 'stateresponse4',
       'stateresponse5', 'stateresponse6', 'stateresponse7', 'protesterdemand1',
       'protesterdemand2', 'protesterdemand3', 'protesterdemand4', 'protest_size_category', 'protester_id_type']))
```

```python
ohe_df.columns
```

```python
ohe_df.head()
```

## Clean the Encoded dataframe

```python
# we want to consolidate each response/demand into their own columns 
# rather than having multiple government response/protester demand columns
ohe_df['arrests'] = ohe_df['stateresponse1_arrests'] + ohe_df['stateresponse2_arrests'] + ohe_df['stateresponse3_arrests'] + ohe_df['stateresponse4_arrests'] + ohe_df['stateresponse5_arrests'] + ohe_df['stateresponse6_arrests'] + ohe_df['stateresponse7_arrests'] 
ohe_df['accomodation'] = ohe_df['stateresponse1_accomodation'] + ohe_df['stateresponse2_accomodation'] + ohe_df['stateresponse3_accomodation'] + ohe_df['stateresponse4_accomodation'] + ohe_df['stateresponse5_accomodation'] + ohe_df['stateresponse6_accomodation'] + ohe_df['stateresponse7_accomodation'] 
ohe_df['beatings'] = ohe_df['stateresponse1_beatings'] + ohe_df['stateresponse2_beatings'] + ohe_df['stateresponse3_beatings'] + ohe_df['stateresponse4_beatings'] + ohe_df['stateresponse5_beatings'] + ohe_df['stateresponse6_beatings'] + ohe_df['stateresponse7_beatings'] 
ohe_df['crowddispersal'] = ohe_df['stateresponse1_crowd dispersal'] + ohe_df['stateresponse2_crowd dispersal'] + ohe_df['stateresponse3_crowd dispersal'] + ohe_df['stateresponse4_crowd dispersal'] + ohe_df['stateresponse5_crowd dispersal'] + ohe_df['stateresponse6_crowd dispersal'] 
ohe_df['ignore'] = ohe_df['stateresponse1_ignore'] + ohe_df['stateresponse2_ignore'] + ohe_df['stateresponse3_ignore']
ohe_df['killings'] = ohe_df['stateresponse1_killings'] + ohe_df['stateresponse2_killings'] + ohe_df['stateresponse3_killings'] + ohe_df['stateresponse4_killings'] + ohe_df['stateresponse5_killings'] + ohe_df['stateresponse6_killings'] + ohe_df['stateresponse7_killings'] 
ohe_df['no_response'] = ohe_df['stateresponse1_None'] + ohe_df['stateresponse2_None'] + ohe_df['stateresponse3_None'] + ohe_df['stateresponse4_None'] + ohe_df['stateresponse5_None'] + ohe_df['stateresponse6_None'] + ohe_df['stateresponse7_None'] 
ohe_df['shootings'] = ohe_df['stateresponse1_shootings'] + ohe_df['stateresponse2_shootings'] + ohe_df['stateresponse3_shootings'] + ohe_df['stateresponse4_shootings'] + ohe_df['stateresponse5_shootings'] 

ohe_df['labor_wage_dispute'] = ohe_df['protesterdemand1_labor wage dispute'] + ohe_df['protesterdemand2_labor wage dispute'] + ohe_df['protesterdemand3_labor wage dispute'] + ohe_df['protesterdemand4_labor wage dispute'] 
ohe_df['land_farm_issue'] = ohe_df['protesterdemand1_land farm issue'] + ohe_df['protesterdemand2_land farm issue'] + ohe_df['protesterdemand3_land farm issue'] + ohe_df['protesterdemand4_land farm issue'] 
ohe_df['no_demands'] = ohe_df['protesterdemand1_None'] + ohe_df['protesterdemand2_None'] + ohe_df['protesterdemand3_None'] + ohe_df['protesterdemand4_None'] 
ohe_df['police_brutality'] = ohe_df['protesterdemand1_police brutality'] + ohe_df['protesterdemand2_police brutality'] + ohe_df['protesterdemand3_police brutality'] + ohe_df['protesterdemand4_police brutality'] 
ohe_df['political_behavior_process'] = ohe_df['protesterdemand1_political behavior, process'] + ohe_df['protesterdemand2_political behavior, process'] + ohe_df['protesterdemand3_political behavior, process'] + ohe_df['protesterdemand4_political behavior, process'] 
ohe_df['price increases_tax_policy'] = ohe_df['protesterdemand1_price increases, tax policy'] + ohe_df['protesterdemand2_price increases, tax policy'] + ohe_df['protesterdemand3_price increases, tax policy'] + ohe_df['protesterdemand4_price increases, tax policy']
ohe_df['removal_of_politician'] = ohe_df['protesterdemand1_removal of politician'] + ohe_df['protesterdemand2_removal of politician'] + ohe_df['protesterdemand3_removal of politician'] + ohe_df['protesterdemand4_removal of politician']
ohe_df['social_restrictions'] = ohe_df['protesterdemand1_social restrictions'] + ohe_df['protesterdemand2_social restrictions'] + ohe_df['protesterdemand3_social restrictions']
ohe_df.head()
```

```python
# turn no_response and no_demands column values to 0 or 1
ohe_df['no_response'] = ohe_df['no_response'].map(lambda b: 1 if b > 0 else 0)
ohe_df['no_demands'] = ohe_df['no_demands'].map(lambda b: 1 if b > 0 else 0)

print(ohe_df['no_demands'].unique()) 
print(ohe_df['no_response'].unique())
```

```python
# clean response columns to (0,1)

ohe_df['arrests'] = ohe_df['arrests'].map(lambda x: 1 if x >= 1 else 0)
ohe_df['accomodation'] = ohe_df['accomodation'].map(lambda x: 1 if x >= 1 else 0)
ohe_df['beatings'] = ohe_df['beatings'].map(lambda x: 1 if x >= 1 else 0)
ohe_df['crowddispersal'] = ohe_df['crowddispersal'].map(lambda x: 1 if x >= 1 else 0)
ohe_df['ignore'] = ohe_df['ignore'].map(lambda x: 1 if x >= 1 else 0)
ohe_df['killings'] = ohe_df['killings'].map(lambda x: 1 if x >= 1 else 0)
ohe_df['no_response'] = ohe_df['no_response'].map(lambda x: 1 if x >= 1 else 0)
ohe_df['shootings'] = ohe_df['shootings'].map(lambda x: 1 if x >= 1 else 0)

ohe_df.head()
```

```python
# same thing for demand columns that have values that are greater than 1
ohe_df['price increases_tax_policy'] = ohe_df['price increases_tax_policy'].map(lambda x: 1 if x >= 1 else 0)
ohe_df['removal_of_politician'] = ohe_df['removal_of_politician'].map(lambda x: 1 if x >= 1 else 0)
```

```python
# we don't need any of these columns 
ohe_df.drop(columns=['no_response', 'no_demands','protesterdemand1_labor wage dispute',
       'protesterdemand1_land farm issue', 'protesterdemand1_None',
       'protesterdemand1_police brutality',
       'protesterdemand1_political behavior, process',
       'protesterdemand1_price increases, tax policy',
       'protesterdemand1_removal of politician',
       'protesterdemand1_social restrictions',
       'protesterdemand2_labor wage dispute',
       'protesterdemand2_land farm issue', 'protesterdemand2_None',
       'protesterdemand2_police brutality',
       'protesterdemand2_political behavior, process',
       'protesterdemand2_price increases, tax policy',
       'protesterdemand2_removal of politician',
       'protesterdemand2_social restrictions',
       'protesterdemand3_labor wage dispute',
       'protesterdemand3_land farm issue', 'protesterdemand3_None',
       'protesterdemand3_police brutality',
       'protesterdemand3_political behavior, process',
       'protesterdemand3_price increases, tax policy',
       'protesterdemand3_removal of politician',
       'protesterdemand3_social restrictions', 'protesterdemand4_.',
       'protesterdemand4_labor wage dispute',
       'protesterdemand4_land farm issue', 'protesterdemand4_None',
       'protesterdemand4_police brutality',
       'protesterdemand4_political behavior, process',
       'protesterdemand4_price increases, tax policy',
       'protesterdemand4_removal of politician','stateresponse1_accomodation', 'stateresponse1_arrests',
       'stateresponse1_beatings', 'stateresponse1_crowd dispersal',
       'stateresponse1_ignore', 'stateresponse1_killings',
       'stateresponse1_None', 'stateresponse1_shootings',
       'stateresponse2_accomodation', 'stateresponse2_arrests',
       'stateresponse2_beatings', 'stateresponse2_crowd dispersal',
       'stateresponse2_ignore', 'stateresponse2_killings',
       'stateresponse2_None', 'stateresponse2_shootings',
       'stateresponse3_accomodation', 'stateresponse3_arrests',
       'stateresponse3_beatings', 'stateresponse3_crowd dispersal',
       'stateresponse3_ignore', 'stateresponse3_killings',
       'stateresponse3_None', 'stateresponse3_shootings',
       'stateresponse4_accomodation', 'stateresponse4_arrests',
       'stateresponse4_beatings', 'stateresponse4_crowd dispersal',
       'stateresponse4_killings', 'stateresponse4_None',
       'stateresponse4_shootings', 'stateresponse5_.',
       'stateresponse5_accomodation', 'stateresponse5_arrests',
       'stateresponse5_beatings', 'stateresponse5_crowd dispersal',
       'stateresponse5_killings', 'stateresponse5_None',
       'stateresponse5_shootings', 'stateresponse6_accomodation',
       'stateresponse6_arrests', 'stateresponse6_beatings',
       'stateresponse6_crowd dispersal', 'stateresponse6_killings',
       'stateresponse6_None', 'stateresponse7_.',
       'stateresponse7_accomodation', 'stateresponse7_arrests',
       'stateresponse7_beatings', 'stateresponse7_killings',
       'stateresponse7_None'], inplace=True)
ohe_df.shape
```
