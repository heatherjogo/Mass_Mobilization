### How does this model compare to what actually happened in these protests?
* 1999 Seattle WTO had **arrests** and **crowd dispersal** (tear gas)
* 2011 Occupy Atlanta was mostly peaceful, but did have **arrests** and **crowd dispersal** with helicopters, SWAT team and motorcycle police.
* 2018 March For Our Lives was peaceful. No arrests, no accomodations, no crowd dispersal, no killings and no shootings or violence. It was **ignored**.
* 2020 Michigan Covid Lockdown was a small protest nicknamed Operation Gridlock. It had no arrests, no beatings, no crowd dispersal, no killings, no shootings or violence.  It was **ignored**.
* 2021 D.C. Riot had **arrests**. No accomodation, but there were **beatings** between law enforcement and the protesters. There was eventually **crowd dispersal**. Five were people **killed**, so there were **shootings**, and a **violent response**.

#### Load model using logistic regression classifiers from pickle files

To improve model performance, this model has dropped all entries in the State Response **ignore** column. The ignore column had a 1 for each time a state ignored a protest, and a zero otherwise. Not including this column creates a much more balanced set of target variables.

```python
# Arrests
with open('../models/04_no_ignore_arrests.pickle', 'rb') as f:
    no_ignore_model_arrests = pickle.load(f)

# Accomodation
with open('../models/04_no_ignore_accomodation.pickle', 'rb') as f:
    no_ignore_model_accomodation = pickle.load(f)

# Beatings
with open('../models/04_no_ignore_beatings.pickle', 'rb') as f:
    no_ignore_model_beatings = pickle.load(f)
    
# Crowd dispersal
with open('../models/04_no_ignore_crowddispersal.pickle', 'rb') as f:
    no_ignore_model_crowd = pickle.load(f)

# Killings
with open('../models/04_no_ignore_killings.pickle', 'rb') as f:
    no_ignore_model_killings = pickle.load(f)

# Shootings
with open('../models/04_no_ignore_shootings.pickle', 'rb') as f:
    no_ignore_model_shootings = pickle.load(f)
    
# Violent response
with open('../models/04_no_ignore_violent_response.pickle', 'rb') as f:
    no_ignore_model_violent = pickle.load(f)
```

```python
predictions_no_ignore_df = pd.DataFrame([], index=X.index)
```

The `violent_response` column is a new feature used by the model. If there was a beating, killing or shooting, then violent_response is set to 1.

```python
# Load model's predictions into a dataframe.
predictions_no_ignore_df['arrests'] = no_ignore_model_arrests.predict(X_sc).tolist()
predictions_no_ignore_df['accomodation'] = no_ignore_model_accomodation.predict(X_sc).tolist()
predictions_no_ignore_df['beatings'] = no_ignore_model_beatings.predict(X_sc).tolist()
predictions_no_ignore_df['crowd_dispersal'] = no_ignore_model_crowd.predict(X_sc).tolist()
predictions_no_ignore_df['killings'] = no_ignore_model_killings.predict(X_sc).tolist()
predictions_no_ignore_df['shootings'] = no_ignore_model_shootings.predict(X_sc).tolist()
predictions_no_ignore_df['violent_response'] = no_ignore_model_violent.predict(X_sc).tolist()

predictions_no_ignore_df
```
